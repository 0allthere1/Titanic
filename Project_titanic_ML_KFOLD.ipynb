```python
def transform_sex(value):
    if value == 'female':
        return 1
    else:
        return 0
```

```python
#SUBSTIITE THE MISSING VALUES FOR MEDIAN 
test["Fare"].fillna(test.groupby("Pclass")["Fare"].transform("mean"),inplace=True)
train2['Sex_binary'] = train2['Sex'].map(transform_sex)
test['Sex_binary'] = test['Sex'].map(transform_sex)
print(test.info())
```

```python
VARIABLES = ['Sex_binary', 'Age']
x = train2[VARIABLES]
y = train2['Survived']

x=x.fillna(-1)
```

```python
# AGE
# Count values
#Change the name to identify "Man", "Woman"


test["Title"]=test["Name"].str.extract("([A-Za-z]+)\.",expand=False)
test['Title']=test['Title'].replace(['Ms','Mlle'],'Miss')
test['Title']=test['Title'].replace(['Mme','Countess','Lady','Dona'],'Mrs')
test['Title']=test['Title'].replace(['Dr','Major','Col','Capt','Sir','Rev','Jonkheer','Don'],'Mr')


#APPLY THE CHANGES

test["Age"].fillna(test.groupby("Title")["Age"].transform("median"),inplace=True)

#DROP
test.drop(columns=['Title'])
```

##CROSS VALIDATION
##VALIDACAO CRUZADAS

```python
#REAMOSTRAGEM

#Deixar a validacao mais confiavel e meus dados mais confiaveis

Dividir os dados em blocos ao inves de serem aleatorios

```

```python
from sklearn.model_selection import KFold

# retorna o indice das linhas de cada divisao
 # n_split = n* de divisoes que queremos no nosso modelo
 # shuffle = usar os dados na ordem que estao ou fazer um sorteio antes de utilizarmos
 # random_state =  random.seed - sementinha do 
```

```python
kf= KFold(n_splits= 3, shuffle= True, random_state= 0)
for train_lines, valid_lines in kf.split(x_falso): 
  #o split vai falar para nos quais linhas vamos usar em cada conjunto de dados
  print("TREINO:", train_lines)
  print("Valid:", valid_lines)
  print()
```

```python
#create a list to save
results=[]

#rodar a mesma validacao porem com a aleatoriedade diferente
#vamos rodar 10 repeticoes com 2 folds

for rep in range(10): #Para o rep no range de 0 a 10 
  print("Rep:", rep) #print
  kf= KFold(n_splits= 2, shuffle= True, random_state= rep)


  for train_lines, valid_lines in kf.split(x): 
    #o split vai falar para nos quais linhas vamos usar em cada conjunto de dados
    print("Treino:", train_lines.shape[0])
    print("Valid:", valid_lines.shape[0])

  #iloc do pandas vai procurar uma funcao e criar um novo dataframe
    x_train, x_valid = x.iloc[train_lines], x.iloc[valid_lines]
    y_train, y_valid = y.iloc[train_lines], y.iloc[valid_lines]
    #print(x_train.head())

  # model
    model = RandomForestClassifier(n_estimators=120, n_jobs=-1, random_state=0)
    model.fit(x_train, y_train)

    p = model.predict(x_valid)

    acc = np.mean(y_valid == p )

    results.append(acc)
    print("Acc:", acc)
    print()
```

```python
np.mean(results)
```
